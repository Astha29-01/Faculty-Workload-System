### OLTP vs OLAP (Simple Meaning)

#### OLTP (Online Transaction Processing)
##### Used for:

- Daily operations
- Insert / Update / Delete
- Fast transactions

##### Example in your project:

- Assign subject to faculty
- Update workload hours
- Add new faculty

#### OLAP (Online Analytical Processing)
##### Used for:

- Reporting
- Analysis
- Dashboard
- Aggregation

##### Example in your project:

- Total workload per department
- Faculty workload trends
- Semester reports

# OLTP Database: faculty_workload_db

#### OLTP Tables
* faculty
* subjects
* workload
##### OLTP Structure
##### faculty(
    faculty_id PK
    faculty_name
    department
)
##### subjects(
    subject_id PK
    subject_name
)
##### workload(
    workload_id PK
    faculty_id FK
    subject_id FK
    hours_assigned
    semester
)

#### OLTP Purpose
- Store real-time transactions
- Maintain data integrity
- Handle updates

### OLAP STAR SCHEMA

* Fact Table: fact_workload

* Dimension Tables:
  - dim_faculty
  - dim_subject
  - dim_semester
  - dim_department


#### FACT TABLE
##### fact_workload(
    workload_id
    faculty_id
    subject_id
    department_id
    semester_id
    hours_assigned
)

* Stores: Numerical measurable data(hours)

#### DIMENSION TABLES

##### dim_faculty(
    faculty_id PK
    faculty_name
)
##### dim_subject(
    subject_id PK
    subject_name
)
##### dim_department(
    department_id PK
    department_name
)
##### dim_semester(
    semester_id PK
    semester_name
)

##### OLTP vs OLAP COMPARISON TABLE(
    Feature	OLTP	OLAP
    Purpose	Transactions	Analytics
    Data	Current	Historical
    Queries	Short	Complex
    Speed	Fast writes	Fast reads
    Schema	Normalized	Star schema
)
Example	Assign workload	Department report

#### OLTP Use Cases
* Add new faculty
* Assign subject

* Delete incorrect record

#### OLAP Use Cases
* Total workload per department
* Semester comparison
* Faculty ranking
* Trend analysis
