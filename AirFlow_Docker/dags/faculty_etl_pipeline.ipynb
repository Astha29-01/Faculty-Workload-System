```python
import pandas as pd
import time
import logging

logging.basicConfig(
    filename="etl_logs.txt",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

# -------------------
# EXTRACT
# -------------------
def extract():
    logging.info("Extract started")
    df = pd.read_csv("../data/input_data.csv")
    logging.info(f"Extracted rows: {len(df)}")
    return df

# -------------------
# TRANSFORM
# -------------------
def transform(df):
    logging.info("Transform started")

    df.fillna(0, inplace=True)

    if "teaching_hours" in df.columns and "admin_hours" in df.columns:
        df["total_workload"] = df["teaching_hours"] + df["admin_hours"]

    logging.info("Transform completed")
    return df

# -------------------
# LOAD
# -------------------
def load(df):
    logging.info("Load started")
    df.to_csv("final_output.csv", index=False)
    logging.info("Data saved to final_output.csv")

# -------------------
# PIPELINE EXECUTION
# -------------------
def run_pipeline(retries=3):
    for attempt in range(retries):
        try:
            df = extract()
            df = transform(df)
            load(df)
            logging.info("Pipeline SUCCESS")
            break

        except Exception as e:
            logging.error(f"Attempt {attempt+1} failed: {e}")
            time.sleep(2)

# RUN
run_pipeline()
```
