```python
import os
import shutil

# Create cloud storage folder
os.makedirs('cloud_storage', exist_ok=True)

# -----------------------------
# Upload to "S3"
# -----------------------------
def upload_to_cloud(local_file):
    filename = os.path.basename(local_file)
    dest = os.path.join('cloud_storage', filename)
    
    shutil.copy(local_file, dest)
    print(f"{filename} uploaded to cloud storage")

# -----------------------------
# Download from "S3"
# -----------------------------
def download_from_cloud(filename):
    src = os.path.join('cloud_storage', filename)
    dest = os.path.join('processed_data', filename)
    
    shutil.copy(src, dest)
    print(f"{filename} downloaded from cloud storage")

# -----------------------------
# Test
# -----------------------------
if __name__ == "__main__":
    upload_to_cloud('../processed_data/processed_faculty.csv')
    download_from_cloud('processed_faculty.csv')
```
