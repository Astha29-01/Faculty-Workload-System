```python
import pandas as pd
import os

print("\n===== HDFS SIMULATION START =====")

# Load dataset
df = pd.read_csv("../data/faculty_workload_raw.csv")

print("\nTotal Rows:", len(df))

# Split dataset into 2 blocks
block1 = df.iloc[:len(df)//2]
block2 = df.iloc[len(df)//2:]

# Save blocks into DataNodes

block1.to_csv(
    "../hdfs_simulation/datanode1/block1.csv",
    index=False
)

block2.to_csv(
    "../hdfs_simulation/datanode2/block2.csv",
    index=False
)

print("\nBlocks Created Successfully")

print("\nBlock 1 → datanode1")
print("Block 2 → datanode2")

print("\n===== HDFS SIMULATION COMPLETE =====")
```

### NameNode
* The NameNode is the master node in Hadoop HDFS.
* It stores metadata about files such as:
  - File names
  - Block locations
  - DataNode mapping

* In this project, the NameNode is simulated using:
hdfs_simulation/namenode folder.

### DataNode
DataNodes store actual data blocks.

In this project:
- datanode1 stores block1.csv
- datanode2 stores block2.csv

This simulates distributed storage.

### File Storage Flow
1. File uploaded to NameNode
2. File divided into blocks
3. Blocks stored across DataNodes
4. Metadata stored in NameNode
