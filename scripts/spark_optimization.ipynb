```python
# ===============================
# TASK 17 — SPARK OPTIMIZATION
# ===============================

from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import time

print("\n===== TASK 17 STARTED =====")

# Create Spark Session
spark = SparkSession.builder \
    .appName("Spark Optimization") \
    .master("local[1]") \
    .getOrCreate()

# Load Dataset
df = spark.read.csv(
    "../data/faculty_workload_raw.csv",
    header=True,
    inferSchema=True
)

print("\nDataset Loaded")
df.show()

# ===============================
# WITHOUT CACHE
# ===============================

print("\nRunning Without Cache...")

start = time.time()

df.groupBy("department").count().show()

end = time.time()

print("Time Without Cache:", end - start)

# ===============================
# WITH CACHE
# ===============================

print("\nApplying Cache...")

df.cache()

start = time.time()

df.groupBy("department").count().show()

end = time.time()

print("Time With Cache:", end - start)

# ===============================
# PARTITIONING
# ===============================

print("\nApplying Repartitioning...")

df_repart = df.repartition(2)

df_repart.groupBy("department").count().show()

# ===============================
# SAVE OUTPUT
# ===============================

pandas_df = df_repart.toPandas()

pandas_df.to_csv(
    "../data/spark_optimized_output.csv",
    index=False
)

print("\nOptimized Output Saved")

spark.stop()

print("\n===== TASK 17 COMPLETED =====")
```
