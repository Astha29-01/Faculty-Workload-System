```python
# ===============================
# TASK 18 — STREAMING SIMULATION
# ===============================

import pandas as pd
import time

print("\n===== TASK 18 STARTED =====")

# Load dataset
df = pd.read_csv("../data/faculty_workload_raw.csv")

print("\nSimulating Streaming...\n")

processed_data = []

# Simulate streaming row-by-row
for index, row in df.iterrows():

    print("Processing Record:")
    print(row)

    # Simulated processing
    hours = row["hours_assigned"]

    # Handle null
    if pd.isna(hours):
        hours = 0

    # Handle negative
    if hours < 0:
        hours = abs(hours)

    row["hours_assigned"] = hours

    processed_data.append(row)

    # Delay to simulate real-time streaming
    time.sleep(1)

print("\nStreaming Completed")

# Convert back to DataFrame
output_df = pd.DataFrame(processed_data)

# Save output
output_df.to_csv(
    "../data/streaming_output.csv",
    index=False
)

print("\nStreaming Output Saved")

print("\n===== TASK 18 COMPLETED =====")
```
