```python
# ===============================
# TASK 16 — SPARK SQL
# ===============================

from pyspark.sql import SparkSession

print("\n===== TASK 16 STARTED =====")

# Create Spark Session
spark = SparkSession.builder \
    .appName("Spark SQL Faculty Workload") \
    .master("local[1]") \
    .getOrCreate()

# Load Dataset
df = spark.read.csv(
    "../data/faculty_workload_raw.csv",
    header=True,
    inferSchema=True
)

print("\nDataset Loaded:")
df.show()

# Create SQL Table
df.createOrReplaceTempView("faculty_workload")

# ===============================
# SQL QUERY 1 — SELECT
# ===============================

print("\nQuery 1 — Faculty from CSE Department")

query1 = spark.sql("""
SELECT faculty_name, subject, hours_assigned
FROM faculty_workload
WHERE department = 'CSE'
""")

query1.show()

# ===============================
# SQL QUERY 2 — GROUP BY
# ===============================

print("\nQuery 2 — Total Hours per Department")

query2 = spark.sql("""
SELECT department,
SUM(hours_assigned) AS total_hours
FROM faculty_workload
GROUP BY department
""")

query2.show()

# ===============================
# SQL QUERY 3 — HAVING
# ===============================

print("\nQuery 3 — Departments with Total Hours > 5")

query3 = spark.sql("""
SELECT department,
SUM(hours_assigned) AS total_hours
FROM faculty_workload
GROUP BY department
HAVING total_hours > 5
""")

query3.show()

# ===============================
# SAVE OUTPUT
# ===============================

pandas_df = query2.toPandas()

pandas_df.to_csv(
    "../data/spark_sql_output.csv",
    index=False
)

print("\nSQL Output Saved Successfully")

spark.stop()

print("\n===== TASK 16 COMPLETED =====")
```
