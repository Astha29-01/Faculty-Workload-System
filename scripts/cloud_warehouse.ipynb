```python
import sqlite3
import pandas as pd

def query_data():
    conn = sqlite3.connect('../processed_data/faculty_lakehouse.db')
    
    query = """
    SELECT department, SUM(hours_assigned) as total_hours
    FROM faculty_workload
    GROUP BY department
    """
    
    df = pd.read_sql(query, conn)
    conn.close()
    
    print("\nDepartment-wise Workload:\n")
    print(df)

if __name__ == "__main__":
    query_data()
```
