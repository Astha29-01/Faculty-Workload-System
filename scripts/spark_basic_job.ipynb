```python
# ===============================
# TASK 14 – SPARK BASIC JOB
# ===============================
import pandas as pandas_df
from pyspark.sql import SparkSession

print("\n===== SPARK JOB STARTED =====")

# Create Spark Session
spark = SparkSession.builder \
    .appName("Faculty Workload Processing") \
    .master("local[1]") \
    .getOrCreate()

# Load CSV Data
df = spark.read.csv(
    "../data/faculty_workload_raw.csv",
    header=True,
    inferSchema=True
)

print("\nLoaded Dataset:")
df.show()

# Process Data
print("\nProcessing Data...")

processed_df = df.withColumn(
    "double_hours",
    df["hours_assigned"] * 2
)

processed_df.show()

# Save Output
pandas_df = processed_df.toPandas()

pandas_df.to_csv(
    "../data/spark_output.csv",
    index=False
)

print("\nOutput Saved Successfully")

spark.stop()

print("\n===== SPARK JOB COMPLETED =====")
```
