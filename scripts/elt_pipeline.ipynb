```python
import pandas as pd
import duckdb
import time

start = time.time()

print("\n--- ELT PIPELINE START ---")

# ========================
# LOAD FIRST
# ========================

con = duckdb.connect("faculty_dw.duckdb")

df = pd.read_csv("../data/faculty_workload_raw.csv")

con.execute("CREATE TABLE raw_data1 AS SELECT * FROM df")

print("\nRaw Data Loaded into DuckDB")

# ========================
# TRANSFORM AFTER LOAD
# ========================

clean_df = con.execute("""

SELECT
faculty_id,
faculty_name,
department,
subject,

ABS(hours_assigned) AS hours_assigned,

semester

FROM raw_data1

""").df()

print("\nTransformed Data:")
print(clean_df)

end = time.time()

print("\nELT Time:", end - start)
```
