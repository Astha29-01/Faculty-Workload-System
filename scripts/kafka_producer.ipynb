```python
# ===============================
# TASK 19 — KAFKA PRODUCER
# ===============================

from kafka import KafkaProducer
import pandas as pd
import json
import time

print("\n===== KAFKA PRODUCER STARTED =====")

# Load dataset
df = pd.read_csv("../data/faculty_workload_raw.csv")

producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

# Send records one-by-one
for _, row in df.iterrows():

    data = row.to_dict()

    producer.send(
        "faculty_topic",
        value=data
    )

    print("Sent:", data)

    time.sleep(1)

producer.flush()

print("\n===== PRODUCER COMPLETED =====")
```
