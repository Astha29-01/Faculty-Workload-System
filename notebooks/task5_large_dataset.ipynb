```python
import pandas as pd
import numpy as np
import time
```

```python
rows = 1_000_000   # 1 Million rows

np.random.seed(42)

df = pd.DataFrame({
    "faculty_id": np.random.randint(1, 100, rows),
    "subject_id": np.random.randint(101, 120, rows),
    "department": np.random.choice(
        ["CSE", "ECE", "IT", "ME"],
        rows
    ),
    "hours_assigned": np.random.randint(
        1,
        6,
        rows
    )
})

print("Dataset Created")

df.head()
```

```python
print("Total Rows:", len(df))

print("\nDepartment Summary:")

dept_summary = df.groupby(
    "department"
)["hours_assigned"].sum()

dept_summary
```

```python
start = time.time()

result = []

for x in df["hours_assigned"]:
    result.append(x * 2)

df["result_loop"] = result

end = time.time()

loop_time = end - start

print("Loop Time:", loop_time)
```

```python
start = time.time()

df["result_vectorized"] = df["hours_assigned"] * 2

end = time.time()

vector_time = end - start

print("Vectorized Time:", vector_time)
```

Vectorized Time << Loop Time
- This proves:
Vectorization is faster

```python
print("\nPerformance Comparison:")

print("Loop Time:", loop_time)

print("Vectorized Time:", vector_time)

if vector_time < loop_time:
    print("\nVectorized method is faster")
```

```python
print("\nMemory Usage Before Optimization:")

df.info(memory_usage="deep")
```

```python
df["faculty_id"] = df["faculty_id"].astype("int16")

df["subject_id"] = df["subject_id"].astype("int16")

df["department"] = df["department"].astype("category")

df["hours_assigned"] = df["hours_assigned"].astype("int8")

print("\nMemory Usage After Optimization:")

df.info(memory_usage="deep")
```

```python
df.to_csv(
    "../processed_data/large_dataset.csv",
    index=False
)

print("Large Dataset Saved")
```
