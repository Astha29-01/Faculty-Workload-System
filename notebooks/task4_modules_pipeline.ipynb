```python
import sys
import os

sys.path.append(
    os.path.abspath("../modules")
)
```

```python
import pandas as pd

from cleaning import (
    remove_duplicates,
    handle_missing_values,
    fix_negative_values,
    drop_null_rows
)

from validation import (
    check_nulls,
    check_duplicates,
    check_negative_values
)

from transformation import (
    merge_datasets,
    add_total_hours,
    department_aggregation
)

print("Modules Loaded Successfully")
```

```python
faculty = pd.read_csv("../raw_data/faculty.csv")

subjects = pd.read_csv("../raw_data/subjects.csv")

workload = pd.read_csv("../raw_data/workload.csv")

print("Data Loaded")
```

```python
print("Null Values:")
print(check_nulls(workload))

print("\nDuplicate Rows:")
print(check_duplicates(workload))

print("\nNegative Values:")
print(check_negative_values(workload))
```

```python
workload = remove_duplicates(workload)

faculty = handle_missing_values(faculty)

workload = fix_negative_values(workload)

workload = drop_null_rows(workload)

print("Data Cleaned")
```

```python
merged_df = merge_datasets(
    workload,
    faculty,
    subjects
)

merged_df = add_total_hours(merged_df)

print("Merged Data:")
merged_df.head()
```

```python
dept_summary = department_aggregation(
    merged_df
)

print("\nDepartment Summary:")
print(dept_summary)
```

```python
merged_df.to_csv(
    "../processed_data/module_output.csv",
    index=False
)

print("Module Output Saved")
```
