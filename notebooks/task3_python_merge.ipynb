```python
import pandas as pd
```

```python
faculty = pd.read_csv("../raw_data/faculty.csv")

subjects = pd.read_csv("../raw_data/subjects.csv")

workload = pd.read_csv("../raw_data/workload.csv")
faculty
```

```python
subjects
```

```python
workload
```

```python
# Remove duplicates
workload = workload.drop_duplicates().copy()

# Fix missing faculty names
faculty["faculty_name"] = faculty["faculty_name"].fillna("Unknown")

# Fix negative hours
workload["hours_assigned"] = workload["hours_assigned"].abs()

# Remove rows with missing hours
workload = workload.dropna()

print("\nCleaned Faculty:")
faculty
```

```python
workload
```

```python
merged_df = workload.merge(
    faculty,
    on="faculty_id"
)

merged_df = merged_df.merge(
    subjects,
    on="subject_id"
)

print("\nMerged Dataset:")
merged_df
```

```python
merged_df.to_csv(
    "../processed_data/merged_workload.csv",
    index=False
)

print("Merged data saved successfully")
```
