```python
import requests
import pandas as pd
```

```python
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)

print("Status Code:", response.status_code)

# Convert JSON to DataFrame
data = response.json()

df = pd.DataFrame(data)

print("\nAPI Data Preview:")
df.head()
```

##### Status Code: 200

##### That means:
* HTTP request successful
* Data transferred successfully

```python
df.to_csv(
    "../raw_data/api_faculty_data.csv",
    index=False
)

print("API Data saved to CSV")
```

```python
print("\nProtocol Used:")

if url.startswith("https"):
    print("Secure HTTPS connection used")
else:
    print("Unsecured HTTP connection")
```

```python
import shutil

source = "../raw_data/api_faculty_data.csv"
destination = "../processed_data/api_faculty_data.csv"

shutil.copy(source, destination)

print("File transferred (FTP simulation)")
```

```python
log_data = {
    "protocol": ["HTTPS"],
    "status": ["Success"],
    "records_received": [len(df)]
}

log_df = pd.DataFrame(log_data)

log_df.to_csv(
    "../logs/network_log.csv",
    index=False
)

print("Network log created")
```

## Networking Explanation

#### HTTP:
HTTP is a protocol used to transfer data between client and server over the internet.

#### HTTPS:
HTTPS is a secure version of HTTP that encrypts data using SSL/TLS.

#### FTP:
FTP (File Transfer Protocol) is used to transfer files between systems.

#### Secure Data Flow:
* Data is encrypted using HTTPS before transmission.
* This protects sensitive information from attackers.

#### Packet Capture:
* Network packets contain request and response data exchanged between systems.
* In real systems, tools like Wireshark are used.
